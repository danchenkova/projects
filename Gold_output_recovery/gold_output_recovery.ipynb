Проверим, что эффективность обогащения рассчитана правильно (rougher.output.recovery). Для подсчета значений объявим функцию, которая будет принимать на вход три параметра: долю золота в концентрате после флотации, долю золота в сырье до флотации и долю золота в отвальных хвостах после флотации.

```python
def recovery_count (clean, feed, tail):
    recovery = (clean* (feed - tail)) / (feed* (clean - tail)) * 100
    return recovery
```

Передадим функции необходимые значения и посчитаем признаки.

```python
train['rougher.recovery_au'] = recovery_count (train['rougher.output.concentrate_au'],
                                              train['rougher.input.feed_au'],
                                              train['rougher.output.tail_au'])
```

```python
print('MAE rougher.output.recovery', mean_absolute_error(train['rougher.output.recovery'][train['rougher.output.recovery'].notnull()], 
                          train['rougher.recovery_au'][train[train['rougher.output.recovery'].notnull()].index]))
```

## Анализ данных

### Исследование концентрации металлов на стадиях очистки

Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки. Для удобства обработки данных выделим нужные нам столбцы в отдельный датафрейм.

```python
me = pd.DataFrame(train, columns = ['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_au',
                                    'rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_au',
                                   'primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb', 'primary_cleaner.output.concentrate_au',
                                    'final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_au'])
me_mean = pd.DataFrame(me.mean())
print(me_mean)
```

Для группировки по металлам и стадиям добавим два признака в датафрейм: 'element' и 'stage'.

```python
element = []
stage = []
for i in range(len(me_mean)):
    element.append(me_mean.index[i][-2:])
    stage.append(me_mean.index[i].partition('.')[0] + '_' + me_mean.index[i].partition('.')[2].partition('.')[0])
    
me_mean['element'] = element
me_mean['stage'] = stage
me_mean.columns = ['mean', 'element', 'stage']

display(me_mean)
```

Чтобы сохранить последовательность стадий (как они следуют друг за другом в реальном производстве), присвоим каждой стадии порядковый номер.

```python
def stage_num (row):
    if row['stage'] == 'rougher_input':
        return 1
    elif row['stage'] == 'rougher_output':
        return 2
    elif row['stage'] == 'primary_cleaner_output':
        return 3
    else:
        return 4

me_mean['stage_number'] = me_mean.apply(stage_num, axis=1)
display(me_mean)
```

Теперь по полученному датафрейму построим сводную таблицу со средними значениями концентрации металлов в сырье на каждой стадии.

```python
display(me_mean.pivot_table(index=['stage_number', 'stage'], columns='element', values='mean', aggfunc='sum'))
```

Из конечной таблицы можно сделать вывод, что на протяжении всего процесса обработки руды концентрация золота и свинца постоянно растет. С серебром другая история: насыщенность сырья серебром после флотации повышается, но вот при последующих этапах очистки металл вымывается из руды.

### Сравнение размеров гранул сырья в выборках

Сравним распределения размеров гранул сырья на обучающей и тестовой выборках.

Построим график для двух распределений.

```python
fig, ax = plt.subplots(figsize=(7,7))

ax.hist(train['rougher.input.feed_size'], density=True, bins=40, alpha=0.4, color='green')
ax.hist(test['rougher.input.feed_size'], density=True, bins=40, alpha=0.3, color='purple')
ax.grid(alpha=0.2, linestyle='--', linewidth=1)
ax.set_title('Распределение гранул сырья', fontsize='large')
ax.set_ylabel('Плотность частоты')
ax.set_xlabel('Размер гранул')
ax.legend(["Train","Test"], fontsize='large', labelspacing=1.3, handlelength=4.5)
```

Внешне графики, вроде, выглядят похожими, только дисперсия для обучающей выборки больше. Рассчитаем значения средних и дисперсий.

```python
print('Среднее, train', train['rougher.input.feed_size'].mean().round(3))
print('Среднее, test', test['rougher.input.feed_size'].mean().round(3))
print('Отношение средних размеров гранул в выборках', (train['rougher.input.feed_size'].mean() / test['rougher.input.feed_size'].mean()).round(3))
```

```python
print('Дисперсия, train: {:.2f}'.format(np.var(train['rougher.input.feed_size'])))
print('Дисперсия, test: {:.2f}'.format(np.var(test['rougher.input.feed_size'])))
print('Отношение дисперсий train/ test: {:.2f}'.format(np.var(train['rougher.input.feed_size']) / np.var(test['rougher.input.feed_size'])))
```

Среднее обучающей выборки смещено вправо относительно тестовой, плюс дисперсия первой больше почти в два раза. 

Проверим равенство средних в двух выборках с помощью p-value. Выдвинем основную (нулевую) гипотезу о том, что средние равны, и между двумя выборками нет значимых отличий. Соответственно, нулевая гипотеза будет заключаться в том, что выборки (а, стало быть, и размеры гранул) отличаются друг от друга. За пороговое значение alpha примем стандартные 0.05.

```python
print(st.ttest_ind(train['rougher.input.feed_size'], test['rougher.input.feed_size'], equal_var=False, nan_policy='omit'), '\n')
if st.ttest_ind(train['rougher.input.feed_size'], test['rougher.input.feed_size'], equal_var=False, nan_policy='omit')[1] < 0.05:
    print('Нулевая гипотеза отвергнута, принята альтернативная')
else:
    print('Нулевая гипотеза не отвергается')
```

Значение p-value получилось очень низким. Мы можем отвергнуть нулевую гипотезу и принять альтернативную. Это означает, что размеры гранул в обучающей и тестовой выборках будем считать в среднем разными.

Это не очень хороший результат, поскольку если распределения сильно отличаются друг от друга, оценка модели может быть неправильной.

### Исследование суммарной концентрации всех веществ на разных стадиях

Рассмотрим, как меняется суммарная концентрация всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. Для этого создадим датафрейм из сумм всех веществ на разных стадиях производства.

```python
train['rougher.input'] = train['rougher.input.feed_ag'] + train['rougher.input.feed_pb'] + train['rougher.input.feed_sol'] + train['rougher.input.feed_au']
train['rougher.output'] = train['rougher.output.concentrate_ag'] + train['rougher.output.concentrate_pb'] + train['rougher.output.concentrate_sol'] + train['rougher.output.concentrate_au']
train['primary_cleaner.output'] = train['primary_cleaner.output.concentrate_ag'] + train['primary_cleaner.output.concentrate_pb'] + train['primary_cleaner.output.concentrate_sol'] + train['primary_cleaner.output.concentrate_au']
train['final.output'] = train['final.output.concentrate_ag'] + train['final.output.concentrate_pb'] + train['final.output.concentrate_sol'] + train['final.output.concentrate_au']

col = ['rougher.input', 'rougher.output', 'primary_cleaner.output', 'final.output']
total_concentrate = pd.DataFrame(train, columns=col)

display(total_concentrate.mean())
```

Концентрация веществ в агрегированном виде между начальной и конечной стадиями выросла, но рост не последователен. Скорее всего, связано с тем, что на разных стадиях концентрация сопутствующих металлов (например, серебра) то растет, то снижается. Это волнообразный процесс.

Посмотрим, как распределяются концентрации веществ на разных стадиях.

```python
fig, ax = plt.subplots(figsize=(9,6))
labels = []
for i in total_concentrate.columns:
    ax.hist(total_concentrate[i], density=True, bins=40, alpha=0.3)
    labels.append(i)
ax.set_title('Распределение суммарной концентрации всех веществ на разных стадиях')
ax.set_xlabel('Суммарная концентрация веществ')
ax.set_ylabel('Плотность частоты')
ax.grid(alpha=0.2, linestyle='--', linewidth=1)
ax.legend(labels, loc='left')    
```

Обнаружены аномальные значения в ряде записей: суммарная концентрация всех веществ на разных этапах обработки равна 0. Также имеется длинный узкий хвост от распределений в сторону нуля. Поскольку все эти показатели вносят искажения в работу модели, удаляем их из обучающей выборки. За порог отсечения примем значение 0.01. Выведем количество записей до удаления и после.

Наблюдается множество небольших положительных и отрицательных корреляций целевых признаков с остальными факторами. Рассчитаем коэффициент ранговой корреляции Спирмена. Используем тот же код, только функции corr() передадим параметр method='spearman'.

```python
display(train.corr(method='spearman').style.background_gradient(cmap='coolwarm'))
```

Ряд связей оказался даже сильнее. Посмотрим, сколько вообще признаков в датасете имеют значение. За порог отсечения примем коэффициент корреляции 0.15, взятый по модулю. А признаки с коэффициентами более 0.40 выведем на экран. Для отбора признаков объявим функцию.

```python
def corr_func(data, series):
    actual_features = []
    for i in data.columns:
        coef_p = series.corr(data[i])
        coef_s = series.corr(data[i], method='spearman')
        if 1 > abs(coef_p) >= 0.15 or 1 > abs(coef_s) >= 0.15:
            actual_features.append(i)
            if 1 > abs(coef_p) >= 0.40 or 1> abs(coef_s) >= 0.40:
                print(i)
    return actual_features
```

Передадим нашей функции набор данных и признак, корреляция с которым нас интересует больше всего. Также посмотрим, какое количество признаков в целом для нас значимо (для прогнозирования каждого таргета).

```python
print('Корреляция с rougher.output.recovery')
columns_rougher = corr_func(train, train['rougher.output.recovery'])
print('Всего признаков', len(columns_rougher), '\n')

print('Корреляция с final.output.recovery')
columns_final = corr_func(train, train['final.output.recovery'])
print('Всего признаков', len(columns_final))
```

secondary_cleaner.state.floatbank4_b_air - т.е. параметры самого процесса очистки.

## Подбор гиперпараметров и обучение моделей

В проекте рассмотрено шесть моделей: линейная регрессия, гребневая регрессия, дерево решений, случайный лес, градиентный бустинг и метод опорных векторов.

Оптимальные гиперпараметры будем подбирать с помощью функции GridSearchCV(). Поскольку перебирать гиперпараметры мы будем не один раз, объявим функцию для подбора этих гиперпараметров. На вход она будет принимать перечень параметров, тип модели, набор обучающих признаков и обучающий целевой признак. 

Возвращать функция будет метрику качества sMAPE (симметричное среднее абсолютное процентное отклонение), которая одинаково учитывает масштаб и целевого признака, и предсказания. Для расчета метрики напишем функцию smape_count и имплементируем ее в функцию подбора параметров модели.

```python
def smape_count (target, prediction):
    sum_diff = 1* ((abs(target - prediction) / ((abs(target) + abs(prediction))/2)).mean())
    return sum_diff
```

```python
select_smape = make_scorer(smape_count, greater_is_better=False)

def select_params(params, model, features, target):
    mod = GridSearchCV(model, params, cv=5, scoring=select_smape)
    %time mod.fit(features, target)
    print('sMAPE:', ((mod.best_score_)*(-1)).round(3))
    print(mod.best_params_)
```

Также объявим функцию, которая будет вычислять общую метрику sMAPE для обоих целевых признаков.

```python
def smape_total (rougher_smape, final_smape):
    return (0.25*rougher_smape + 0.75*final_smape).round(3)
```

### Оценка моделей

Оценим итоги по sMAPE и в разрезе прочих метрик. Объявим функцию для подсчета основных метрик для моделей.

```python
def compare_table(models, features, target):
    df = pd.DataFrame(columns=['models', 'sMAPE', 'MSE', 'RMSE', 'R2'])
    for name, model in models.items():
        prediction = model.predict(features)
        smape = smape_count (target, prediction)
        mse = mean_squared_error(target, prediction).round(3)
        rmse = (mse ** 0.5).round(3)
        r2 = r2_score(target, prediction).round(3)
        df = df.append({'models': name, 'sMAPE': smape, 'MSE': mse, 'RMSE': rmse, 'R2': r2}, ignore_index=True)
    return df
```

Создадим списки моделей для предсказания rougher.output.recovery, final.output.recovery и применим функцию для расчета метрик на тренировочной выборке.

Между целевым признаком rougher.output.recovery установлена наиболее сильная корреляция с количеством добавляемых во время очистки реагентов (сульфида натрия) и объемом воздуха. Второй целевой признак final.output.recovery более всего коррелирован с первым - rougher.output.recovery. Имеется ряд сильных зависимостей между прогнозирующими факторами, что может сделать результаты неустойчивыми.

При проверке моделей на тестовой выборе лучше всех проявила себя комбинация из метода опорных векторов и градиентного бустинга. На основе предсказанных моделями величин получено итоговое отклонение sMAPE (0.084). Значение сопоставлено со sMAPE константной модели (0.089), предсказывающей медиану. Ошибка выбранных моделей меньше ошибки дамми-модели, что говорит о более надежных предсказаниях.

Для дальнейшего улучшения качества прогнозирования имеет смысл провести перебалансировку выборок с целью равномерного распределния в обучающем и тестовом наборах данных сырьевых гранул разных размеров.
